filling event na vlaues with no events and event type with none

```python
calender['event_name_1'] = calender['event_name_1'].fillna('NoEvent')
calender['event_type_1'] = calender['event_type_1'].fillna('None')
calender['event_name_2'] = calender['event_name_2'].fillna('NoEvent')
calender['event_type_2'] = calender['event_type_2'].fillna('None')
```

```python
calender.isna().sum()
```

Duplicate checks fpr d column

```python
dup_calender = calender[calender.duplicated(subset='d')]
dup_calender
```

```python
dup_sales = sales[sales.duplicated(subset='id')]
dup_sales
```

No duplicates found

date alignment

```python
sales_days = [col for col in sales.columns if col.startswith('d_')]
calendar_days = calender['d'].tolist()

missing_in_calendar = set(sales_days) - set(calendar_days)
missing_in_sales = set(calendar_days) - set(sales_days)

print(f"Total sales day columns: {len(sales_days)}")
print(f"Total calendar day rows: {len(calendar_days)}")
print("Days missing in calendar:", missing_in_calendar)
print("Days missing in sales:", missing_in_sales, "\n")
```

Checking for invalid or extreme sales values

```python
sales_values = sales[sales_days]
min_val = sales_values.min().min()
max_val = sales_values.max().max()
print(f"Min sales value: {min_val}")
print(f"Max sales value: {max_val}")

if min_val < 0:
    print("Warning: Negative sales found!")
if max_val > 10000:
    print("Warning: Some sales values seem abnormally large (>10,000)")
```
